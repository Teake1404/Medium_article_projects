```python
import pandas as pd
import numpy as np
import os 
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv([f for f in os.listdir(os.getcwd()) if f.endswith('.csv')][1])
```

```python
hit_rate=df['top_20'].mean()
```

```python
def rate_bucket(rate):
    if rate < 0.02:
        return 'low'
    elif rate < 0.05:
        return 'medium'
    else:
        return 'high'

df['rate_bucket'] = df['CTR'].apply(rate_bucket)
```

```python
p_signal_given_hit=df[df['top_20']]['rate_bucket'].value_counts(normalize=True)

p_signal_given_not_hit=df[~df['top_20']]['rate_bucket'].value_counts(normalize=True)
```

```python
p_signal_given_hit
```

```python
p_signal_given_not_hit
```

```python
def produce_posterior_p(p,not_p):
    P_signal_given_hit = p
    P_signal_given_not_hit = not_p
    
    P_signal = (P_signal_given_hit * hit_rate) + (P_signal_given_not_hit * (1 - hit_rate))

    #Whatâ€™s the probability a video is a hit, given an early signal (like high CTR)?
    P_hit_given_signal = (P_signal_given_hit * hit_rate) / P_signal
    return round(P_hit_given_signal,3)
        
```

```python
P_hit_given_signal_high=produce_posterior_p(p_signal_given_hit['high'],p_signal_given_not_hit['high'])
P_hit_given_signal_medium=produce_posterior_p(p_signal_given_hit['medium'],p_signal_given_not_hit['medium'])
```

```python
def match_poterior_p_rate_bucket(rate_bucket):
    if rate_bucket == 'medium':
        return P_hit_given_signal_medium
    elif rate_bucket == 'high':
        return P_hit_given_signal_high
```

```python
df['posterior_P_hit']=df['rate_bucket'].apply(match_poterior_p_rate_bucket)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(
    x=df['rate_bucket'],
    y=df['posterior_P_hit'],
    order=['low', 'medium', 'high'],
    palette='viridis'
)
plt.title("Posterior P(Hit) by Like Rate Bucket")
plt.ylabel("Posterior Probability of Hit")
plt.xlabel("Like Rate Bucket")
plt.ylim(0, 1)
plt.show()
```

```python
df_sorted = df.sort_values('posterior_P_hit', ascending=False).reset_index()
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df_sorted, x=df_sorted.index, y='posterior_P_hit', hue='top_20', palette='coolwarm')
plt.title("Posterior P(Hit) per Post (sorted)")
plt.xlabel("Post Index (sorted by predicted likelihood)")
plt.ylabel("Posterior Probability of Hit")
plt.legend(title="Actual Hit")
plt.grid(True)
plt.show()
```
